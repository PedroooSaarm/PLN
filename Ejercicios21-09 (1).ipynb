```python
import spacy

nlp = spacy.load("es_core_news_sm")
with open("C:\\Users\\Pedro\\Desktop\\gabriel_garcia_marquez_cien_annos_soledad.txt", "r", encoding="utf-8") as texto:
    novela = texto.read()

doc = nlp(novela)

palabras_unicas = {}
apariciones = {}

for token in doc:
    if not token.is_stop and token.is_alpha:
        palabra = token.text.lower()
        if palabra in palabras_unicas:
            palabras_unicas[palabra] += 1
        else:
            palabras_unicas[palabra] = 1

contador = sum(palabras_unicas.values())
print(f"Número de palabras en la novela: {contador}")

numero_palabras_unicas = len(palabras_unicas)
print(f"Número de palabras únicas en la novela: {numero_palabras_unicas}")

macondo = palabras_unicas.get('macondo', 0)
print(f"Número de veces que aparece '{'macondo'}' en la novela: {macondo}")

palabras_mas_frecuentes = sorted(palabras_unicas.items(), key=lambda x: x[1], reverse=True)[:100]
print("Las 100 palabras más frecuentes:")
for palabra, frecuencia in palabras_mas_frecuentes: 
    print(f"{palabra}: {frecuencia}")
```

```python
#Bigrama
import numpy as np
import random

# Carga el fichero de texto
text = novela

# Aquí están todos los caracteres únicos que aparecen en este texto
chars = sorted(list(set(text)))
vocab_size = len(chars)

# crea un mapeo de caracteres a enteros
stoi = { ch:i for i,ch in enumerate(chars) }
itos = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [stoi[c] for c in s] # encoder: toma una cadena, devuelve una lista de enteros
decode = lambda l: ''.join([itos[i] for i in l]) # decoder: toma una lista de enteros, devuelve una cadena

# Preparación de la tabla que contendrá los bigramas
bigram_table = np.zeros((vocab_size, vocab_size))

# Recorre el texto y cuenta los bigramas
for i in range(len(text)-1):
    bigram_table[stoi[text[i]], stoi[text[i+1]]] += 1

# Normaliza las filas de la tabla
bigram_table = bigram_table / bigram_table.sum(axis=1, keepdims=True)


def generate_text(bigram_table, n_words, word="H"):
    text = [word]
    key_vocab = list(range(vocab_size))
    for i in range(n_words):
        key = random.choices(key_vocab, bigram_table[stoi[text[-1]]])[0]
        word = itos[key]
        text.append(word)
    return ''.join(text)


generate_text(bigram_table, 100, "C")
```

```python
#Unigrama

# Aquí están todos los caracteres únicos que aparecen en este texto
chars = sorted(list(set(text)))
vocab_size = len(chars)

# crea un mapeo de caracteres a enteros
stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}

# Preparación de la tabla que contendrá las probabilidades de unigramas
unigram_table = np.zeros(vocab_size)

# Recorre el texto y cuenta los unigramas
for i in text:
    unigram_table[stoi[i]] += 1

# Normaliza las filas de la tabla
unigram_table = unigram_table / unigram_table.sum(axis=0, keepdims=True)

def generate_text(unigram_table, n_words, word="H"):
    text = [word]
    key_vocab = list(range(vocab_size))

    for i in range(n_words):
        key = random.choices(key_vocab, bigram_table[stoi[text[-1]]])[0]
        word = itos[key]
        text.append(word)
    return ''.join(text)

generated_text = generate_text(unigram_table, 100, "C")

print(generated_text)


```

```python
# Trigrama
# Aquí están todos los caracteres únicos que aparecen en este texto
chars = sorted(list(set(text)))
vocab_size = len(chars)

# crea un mapeo de caracteres a enteros
stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}

# Preparación de la tabla que contendrá las probabilidades de trigramas
trigram_table = np.zeros((vocab_size, vocab_size, vocab_size))

# Recorre el texto y cuenta los trigramas
for i in range(len(text) - 2):
    trigram_table[stoi[text[i]], stoi[text[i + 1]], stoi[text[i + 2]]] += 1

# Normaliza las matrices de trigramas
trigram_table /= np.sum(trigram_table, axis=(1, 2), keepdims=True)

def generate_text(trigram_table, n_words, word=("H", "o")):
    text = list(word)
    key_vocab = list(range(vocab_size))
    for i in range(n_words - len(word)):
        key = random.choices(key_vocab, trigram_table[stoi[text[-2]], stoi[text[-1]], :])[0]
        word = itos[key]
        text.append(word)
    return ''.join(text)

generated_text = generate_text(trigram_table, 100, word=("C", "o"))

print(generated_text)
```

```python
# Calcula la perplejidad del modelo de lenguaje basado en bigram_table.

def perplexity(bigram_table, text):
    perplexity = 0
    for i in range(len(text)-1):
        c1 = stoi[text[i]]
        c2 = stoi[text[i+1]]
        probability = bigram_table[c1][c2]
        if probability == 0:
            probability = 1e-10
        perplexity += np.log(probability)
    perplexity = np.exp(-perplexity/len(text))
    return perplexity

def perplexity_unigram(unigram_table, text):
    perplexity = 0
    for i in range(len(text)-1):
        c1 = stoi[text[i]]
        probability = unigram_table[c1]
        if probability == 0:
            probability = 1e-10
        perplexity += np.log(probability)
    perplexity = np.exp(-perplexity/len(text))
    return perplexity


def perplexity_trigram(trigram_table, text):
    perplexity = 0
    for i in range(len(text)-2):
        c1 = stoi[text[i]]
        c2 = stoi[text[i+1]]
        c3 = stoi[text[i+2]]
        probability = trigram_table[c1][c2][c3]
        if probability == 0:
            probability = 1e-10
        perplexity += np.log(probability)
    perplexity = np.exp(-perplexity/len(text))
    return perplexity

txt1= "Muchos años después, frente al pelotón de fusilamiento, el coronel Aureliano Buendía"

print("Perplejidad bigrama: ")
print(perplexity(bigram_table, txt1))
print(perplexity(bigram_table, txt1[::-1]))

print("Perplejidad unigrama: ")
print(perplexity_unigram(unigram_table, txt1))
print(perplexity_unigram(unigram_table, txt1[::-1]))

print("Perplejidad trigrama: ")
print(perplexity_trigram(trigram_table, txt1))
print(perplexity_trigram(trigram_table, txt1[::-1]))

```
