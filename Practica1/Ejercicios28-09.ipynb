```python
import docx
from string import digits
import spacy

nlp = spacy.load("es_core_news_sm")
# Procesar un texto con el modelo de spaCy

def eliminar_stopwords(texto, stopwords):
    return ' '.join([word for word in texto.split(' ') if word not in stopwords])
# Abre el archivo DOCX
archivos = ['01','02','03','04','05','06','07','08','09','10','11','12']
for i in archivos:
    texto=''
    documento = docx.Document(fr'C:\Users\Pedro\Desktop\Textos\doc{i}.docx')
    for paragraph in documento.paragraphs:
        texto += paragraph.text
        texto+=" "
    texto = texto.lower().replace(".", "").replace(",", "").replace('"', "").replace("  ", " ").replace(":","")
    texto = ''.join([i for i in texto if not i.isdigit()])
    texto=texto.replace("  ", " ")
    doc = nlp(texto)
    stopwords = []
    # Iterar sobre los tokens en el documento procesado
    for token in doc:
        if token.is_stop == True:
            token = token.text
            stopwords.append(token)
    texto = eliminar_stopwords(texto, stopwords)
    tokens = texto.split()

    locals()["texto_"+i] = texto
    locals()["tokens_"+i] = tokens
```
